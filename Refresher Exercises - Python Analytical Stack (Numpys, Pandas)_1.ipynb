### swapping dimensions
We can also change the position of dimensions, in several ways.

```python
# For example, we can circularly shift dimensions using np.roll
array3D = np.ones((4, 10, 20))
array3D.shape
```

### concatenation
First, we perform concatenation.
This is assembling of two 2D arrays into a larger 2D array. 

```python
# first, we try to concatenate array1 and array2
np.concatenate([array1, array2])
```

```python
# However, this gives an error due to different dimensions
```

```python
# We can therefore only concatenate along the second dimension
array_conc = np.concatenate([array1, array2], axis = 1)
array_conc
```

```python
# Success
```

### Broadcasting

The element with less dimensions is "broadcasted" across the array1.

```python
# For example, the simple case of a scalar time a matrix
2 * array1
```

```python
# Here, we see the result which is simple multiplication
```

```python
# When multiple arrays are broadcast, elements get repeated.
```

```python
# And then we create a 2D array
array2D = np.ones((6,4))
array2D
```

```python
# And here is the result
array1D * array2D
```

```python
# Here array1D got broadcasted over each line of the 2D array array2D.
```
